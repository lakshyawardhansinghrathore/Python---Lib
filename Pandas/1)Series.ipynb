# Pandas Series

A Series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.

## Creating Series
series -> in vertically
There are multiple ways to create a pandas Series:

```python
import numpy as np 
import pandas as pd 
```

```python
labels = ['a', 'b', 'c'] #indexes list of string type
my_list = [10,20,30] # list of int type
arr = np.array([10,20,30]) # array
d = {'a':10,'b':20,'c':30} #dictionary
```

```python
pd.Series(my_list) # 0,1,2 in outputs which we know indexes are called labels
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr) #dont pass array with two dimensions as it will give error as series are one dimensionals
```

```python
pd.Series(d)
```
