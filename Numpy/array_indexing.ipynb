**NUMPY INDEXING AND SLICING OF MATRIX**

```python
arr = np.arange(1,31).reshape(6,5) 
arr
```

```python
arr[1] # Indexing on rows
```

```python
arr[0,0]
```

```python
arr[1][2]
```

```python
arr[0:2,1:3]
```

```python
slice = arr[3:,3:]
slice
```

```python
arr[:,2]
```

**BOOLEAN INDEXING**

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr %2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
